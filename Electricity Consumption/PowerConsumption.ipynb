```python
import pandas as pd

df=pd.read_csv(r"C:\Users\kiran\OneDrive\Documents\Projects\Morocco Electricity Consumption\Electricity+Consumption\powerconsumption.csv")
```

```python
df.head()
```

```python
#Read in 4 columns from the powerconsumption.csv file and cast 'Datetime' as a datetime datatype.

#Create a 'total_consumption' column which represents the sum of the three 'PowerConsumption' columns.


#Create two columns by extracting the dateparts from the 'Datetime' column.

df["total_consumption"]=df["PowerConsumption_Zone1"]+df["PowerConsumption_Zone2"]+df["PowerConsumption_Zone3"]

#powerConsumption=df[["Datetime",]
```

```python
df["Datetime"] = df["Datetime"].astype("datetime64[ns]")
```

```python
df["hour"]= df["Datetime"].dt.hour
```

```python
df["DayOfWeek"]= df["Datetime"].dt.dayofweek
```

```python
df.head()
```

```python
powerConsumption = df[["Datetime","PowerConsumption_Zone1","PowerConsumption_Zone2","PowerConsumption_Zone3","total_consumption","DayOfWeek","hour"]]
```

```python
powerConsumption.head()
```

```python
powerConsumption.set_index("Datetime")
```

```python
#Resample the data into 1 hour increments.
#Filter the data down to the month of January 2017.
#Create a stacked line chart using the 'PowerConsumption' columns.
sample_powerConsumption=(powerConsumption
 .set_index("Datetime")
 .resample("H")
 ["PowerConsumption_Zone1","PowerConsumption_Zone2","PowerConsumption_Zone3"]
 .mean()
 .loc["2017-01"]

)


```

```python
sample_powerConsumption.head()
```

```python
#Create a stacked line chart using the 'PowerConsumption' columns.
import seaborn as sns
sample_powerConsumption.plot.area(title="Power consumption by zone JAn 2017",xlabel ="Date",ylabel="consumption in KW hours")
sns.despine()
```

```python
day_of_week=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
```

```python
powerConsumption["day_name"]=pd.Categorical(powerConsumption["Datetime"].dt.day_name(),categories= day_of_week)

#powerConsumption["day_name"]=powerConsumption["Datetime"].dt.day_name()
```

```python



sns.heatmap(
powerConsumption.pivot_table(index="day_name",
                             columns="hour",
                             values="total_consumption",
                             aggfunc="mean"
   
   )
).set(title="Power consumption by day of week")
```
